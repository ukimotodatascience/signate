```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeRegressor
```

```python
df_train = pd.read_csv('train.csv')
df_train.head()
```

```python
df_test = pd.read_csv('test.csv')
df_test.head()
```

```python
df_train.isnull().sum()
```

```python
df_test.isnull().sum()
```

```python
train_x = df_train[['age','sex','bmi','children','smoker','region']]
train_t = df_train['charges']
```

```python
train_x['sex'] = train_x['sex'].replace('male', '0').replace('female','1')
df_test['sex'] = df_test['sex'].replace('male','0').replace('female','1')
```

```python
train_x['smoker'] = train_x['smoker'].replace('no', '0').replace('yes','1')
df_test['smoker'] = df_test['smoker'].replace('no','0').replace('yes','1')
```

```python
train_x['region'] = train_x['region'].replace('southeast', '0').replace('southwest','1').replace('northwest','2').replace('northeast','3')
df_test['region'] = df_test['region'].replace('southeast','0').replace('southwest','1').replace('northwest','2').replace('northeast','3')
```

```python
#train_t = train_t.replace('0', '低').replace('1','中').replace('2','高')
```

```python
train_t.unique()
```

```python
test_x = df_test[['age','sex','bmi','children','smoker','region']]
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(train_x)
x_test_scaled = scaler.fit_transform(test_x)
```

```python
model = DecisionTreeRegressor(max_depth=5,random_state=0)
model.fit(train_x,train_t)
```

```python
df_test['charges'] = model.predict(x_test_scaled)
```
